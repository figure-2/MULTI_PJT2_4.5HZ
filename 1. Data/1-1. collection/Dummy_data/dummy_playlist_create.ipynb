```python
# null 값 확인 코드
import pandas as pd
import random

file_path = '/home/ubuntu/project/4.5HZ/Data/완성/최종완성_적재_트랙(12월20일).csv'

df = pd.read_csv(file_path, encoding='utf-8')

df.info()
df.isnull().sum()
```

-------------------------------------------------------------------------------------------------------------------------
# 팀원 장르 지정 플레이리스트
-------------------------------------------------------------------------------------------------------------------------

```python
# 장르를 기준으로 1차 필터링 후 랜덤으로 뽑는 코드 // 팀원 장르 지정

genres = [ 'edm', 'rock', 'k-pop', 'pop', 'r-n-b']
# genres = [ 'acoustic','dance', 'edm', 'hip-hop','k-pop', 'pop', 'r-n-b', 'rock', 'romance','soul']

playlist_samples = []
playlist_id_counter = 1
user_id_counter = 1

for genre in genres:
    genre_rows = df[df['genres'].str.contains(genre)]
    genre_samples = [genre_rows.sample(500) for _ in range(1)] 
    for idx, sample in enumerate(genre_samples, start=1):
        sample['playlist_id'] = playlist_id_counter
        playlist_id_counter += 1
        sample['USER_ID'] = user_id_counter
        user_id_counter += 1
        sample['id'] = ''
        sample['rating'] = [round(random.uniform(0, 5)) for _ in range(len(sample))]
        sample['cnt'] = [random.randint(1, 20) if rating == 0
                              else random.randint(21, 40) if rating == 1
                              else random.randint(41, 60) if rating == 2
                              else random.randint(61, 80) if rating == 3
                              else random.randint(81, 100) if rating == 4
                              else random.randint(101, 120) for rating in sample['rating']]
        
        # sample['playlist_id'] = '' 
        # sample['user_id'] = '' 
        
    playlist_samples.extend(genre_samples)

playlist_all_1 = pd.concat(playlist_samples, ignore_index=True)

#playlist = playlist_all[['playlist_id', 'USER_ID']]
```

```python
playlist_all_1.info()
```

```python
playlist = playlist_all_1[['playlist_id','USER_ID' ]] # , 'genres', 'release_date'
playlist.to_csv(f'/home/ubuntu/project/4.5HZ/Data/팀원_genre_ramdom_playlist(12월26일).csv', index=False)
```

```python
playlist_all_1.rename(columns={'playlist_id' : 'PLAYLIST_ID'}, inplace = True )
playlist_all_1.rename(columns={'track_id' : 'TRACK_ID'}, inplace = True )


playlist_1 = playlist_all_1[['PLAYLIST_ID','TRACK_ID', 'cnt', 'rating']]
playlist_1.to_csv(f'/home/ubuntu/project/4.5HZ/Data/팀원_genre_ramdom_playlist_track(12월26일).csv', index=False)
```

# 장르별 플레이리스트
-------------------------------------------------------------------------------------------------------------------------

```python
import pandas as pd
import random

genres = ['acoustic','dance', 'edm', 'hip-hop','k-pop', 'pop', 'r-n-b', 'rock', 'romance','soul']

playlist_samples = []
playlist_id_counter = 6
user_id_counter = 6

for genre in genres:
    genre_rows = df[df['genres'].str.contains(genre)]
    for i in range(500):  # 플레이리스트 500개 만들기
        sample = genre_rows.sample(500)  # 플레이리스트 당 노래 500곡
        sample['playlist_id'] = playlist_id_counter
        playlist_id_counter += 1
        sample['USER_ID'] = user_id_counter
        user_id_counter += 1
        sample['id'] = ''
        sample['rating'] = [round(random.uniform(0, 5)) for _ in range(len(sample))]
        sample['cnt'] = [random.randint(1, 20) if rating == 0
                              else random.randint(21, 40) if rating == 1
                              else random.randint(41, 60) if rating == 2
                              else random.randint(61, 80) if rating == 3
                              else random.randint(81, 100) if rating == 4
                              else random.randint(101, 120) for rating in sample['rating']]
        playlist_samples.append(sample)

playlist_all_3 = pd.concat(playlist_samples, ignore_index=True)
```
