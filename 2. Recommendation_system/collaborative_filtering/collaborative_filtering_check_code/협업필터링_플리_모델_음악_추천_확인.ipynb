```python
import pandas as pd
import pickle

# 데이터 로딩
file_path = '/home/ubuntu/project/4.5HZ/Data/완성/플레이리스트/플레이리스트_협업필터링(12월26일).csv'
df = pd.read_csv(file_path, encoding='utf-8')

# 모델 불러오기
loaded_model = pickle.load(open('/home/ubuntu/project/4.5HZ/추천시스템/협업필터링/Collaborative_Filtering_model.pkl', 'rb'))
# Collaborative_Filtering_model.pkl 파일로 저장된 학습된 협업 필터링 모델을 불러옵니다.

# 접속한 사용자의 플레이리스트에 있는 노래 및 평점 (예시로 user_id = 8 사용)
user_id = 5 # 1~10005 for문
user_tracks = df[df['TRACK_ID'] == user_id][['TRACK_ID', 'rating']]
# 사용자가 평가한 플레이리스트 중에서 TRACK_ID가 user_id인 행을 추출합니다.

# 모든 노래에 대한 추천 예측
all_tracks = df['TRACK_ID'].unique()
predictions = [loaded_model.predict(user_id, track_id) for track_id in all_tracks]
# 사용자가 평가한 플레이리스트와 모든 노래에 대해 협업 필터링 모델을 사용하여 예측을 수행합니다.

# 사용자가 좋아할 만한 음악 20개 추천
predictions.sort(key=lambda prediction: prediction.est, reverse=True)
top_20_recommendations = [(prediction.iid, prediction.est) for prediction in predictions[:20]]
# 예측 결과를 평점(est)에 따라 내림차순으로 정렬하고 상위 20개를 추출합니다.

# 상위 20개의 노래에 대한 예측 평점 출력
print("상위 20개 노래에 대한 예측 평점:")
for track_id, est_rating in top_20_recommendations:
    print(f"Track ID: {track_id}, 예상 평점: {est_rating}")


# 20개의 추천 음악 리스트 출력 및 cnt * rating 값 계산
top_20_with_score = []
for track_id, est_rating in top_20_recommendations:
    track_info = df[df['TRACK_ID'] == track_id]
    cnt = track_info['cnt'].values[0]
    rating = track_info['rating'].values[0]
    score = cnt * rating
    top_20_with_score.append((track_id, est_rating, cnt, rating, score))
# 상위 20개의 추천 음악에 대해 각 음악의 cnt와 rating을 추출하고 cnt * rating 값을 계산합니다.

# 가장 큰 cnt * rating 값을 가진 5개 추천 음악 추출
top_20_with_score.sort(key=lambda x: x[4], reverse=True)
top_5_highest_score = top_20_with_score[:5]
# cnt * rating 값에 따라 내림차순으로 정렬하고 상위 5개를 추출합니다.

# 20개의 추천 음악 리스트 출력 (cnt와 rating을 함께 출력하고 높은 순서로 정렬)
# print("20개 추천 음악 리스트 :")
# for track_id, est_rating, cnt, rating, score in top_20_with_score:
#     print(f"Track ID: {track_id}, 예상 평점: {est_rating}, cnt: {cnt}, rating: {rating}, cnt * rating: {score}")

# 가장 큰 cnt * rating 값을 가진 5개 추천 음악 출력
print("\n최종 5개 추천 음악:")
for track_id, est_rating, cnt, rating, score in top_5_highest_score:
    print(f"Track ID: {track_id}, 예상 평점: {est_rating}, cnt: {cnt}, rating: {rating}, cnt * rating: {score}")
    #print(f"Track ID: {track_id}, 예상 평점: {est_rating})
# 상위 20개의 추천 음악과 상위 5개의 추천 음악을 출력합니다.
```
